```python
import numpy as np
import matplotlib.pyplot as plt
im=plt.imread('ata.jpeg')
type(im)
im.shape
```

```python
plt.imshow(im)
plt.show()
```

```python
m,n,p = im.shape #Assigns location and size of the picture to m, n and p by order.
m,n,p
```

```python
#Converts the picture black and white.

new_im=np.zeros((m,n),dtype=float) 

for i in range (m):
    for j in range (n):
        s = (im[i,j,0]+im[i,j,1]+im[i,j,2])/3
        new_im[i,j] = s

plt.imshow(new_im,cmap='gray')
plt.show() 
plt.imsave('ata_gray.png',new_im,cmap='gray')
```

```python
#Rotates the picture 90 degrees left.

new_im=np.zeros((n,m),dtype=float) 

for i in range (m):
    for j in range (n):
        s = (im[i,j,0]+im[i,j,1]+im[i,j,2])/3
        new_im[j,i] = s

plt.imshow(new_im,cmap='gray')
plt.show() 
plt.imsave('left.png',new_im,cmap='gray')
```
